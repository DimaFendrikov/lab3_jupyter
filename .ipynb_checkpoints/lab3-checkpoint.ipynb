```python
import pandas as pd
import numpy as np

file_path = 'En_In.xls'
Energy = pd.read_excel(file_path, skiprows=16, skipfooter=38)

Energy = Energy.iloc[:, 2:]

Energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

Energy['Energy Supply'] = pd.to_numeric(Energy['Energy Supply'], errors='coerce')
Energy['Energy Supply'] = Energy['Energy Supply'] * 1_000_000
  
Energy['Country'] = Energy['Country'].str.replace(r"\s*\(.*\)|\d+", "", regex=True)

rename_Country_Energy = {
    "United States of America": "United States",
    "Republic of Korea": "South Korea",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
}

Energy['Country'] = Energy['Country'].replace(rename_Country_Energy)

Energy.loc[Energy['Country'].isin(['American Samoa','South Korea','Bolivia'])]
```

```python
file_path = 'gpd.csv'
GPD = pd.read_csv(file_path, skiprows=4) 

GPD.rename(columns={'Country Name': 'Country'}, inplace=True)

rename_Country_GPD = {
    "Korea, Rep.": "South Korea",
    "Iran, Islamic Rep.": "Iran",
    "Hong Kong SAR, China": "Hong Kong"
}

GPD['Country'] = GPD['Country'].replace(rename_Country_GPD)
GPD.head(1)
```

```python
file_path = 'scimagojr.xlsx'
scimagojr = pd.read_excel(file_path)

rank_top_15 = scimagojr[scimagojr['Rank'] <= 15]

Energy_data = pd.merge(rank_top_15, Energy, on='Country', how='inner')

Result = pd.merge(Energy_data, GPD, on='Country', how='inner')

Result.set_index('Country', inplace=True)

columns = [
    'Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations',
    'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita',
    '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015'
]
Result = Result[columns]


Result.head(3)
# Result.shape
```

```python
def task_eight():
    years = ['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    
    Result['avgGPD'] = Result[years].mean(axis=1)
    
    sorted_dataframe = Result.sort_values('avgGPD', ascending=False)
    
    countries = sorted_dataframe['avgGPD'].head(15)
    
    return countries

task_eight()
```

```python
def task_nine():

    avg_GPD_countries = task_eight()

    avg_5_country = avg_GPD_countries.index[4]

    GPD_2006 = Result.loc[avg_5_country, '2006']
    GPD_2015 = Result.loc[avg_5_country, '2015']
    GPD_change = GPD_2015 - GPD_2006

    return (avg_5_country, GPD_change)
    
task_nine()
```

```python
def task_ten():

    max_Renewable_row = Result['% Renewable'].idxmax()
    max_Renewable_value = Result.loc[max_Renewable_row, '% Renewable'] 
    
    return (max_Renewable_row, max_Renewable_value)

task_ten()
```

```python
def task_eleven():

    Result['Population'] = Result['Energy Supply'] / Result['Energy Supply per Capita']
    sorted_result = Result.sort_values('Population', ascending=False)
    sixth_country = sorted_result.iloc[5]
    
    return (sixth_country.name, sixth_country['Population'])

task_eleven()
```

```python
def task_twelve():

    Result['Citations documents per person'] = Result['Citable documents'] / Result['Population']

    answer = Result['Citations documents per person'].corr(Result['Energy Supply per Capita'])

    return answer

task_twelve()
```

```python
def task_thirteen():

    median_renewable = Result['% Renewable'].median()
    
    Result['Renewable > Median'] = (Result['% Renewable'] >= median_renewable).astype(int)
    
    sorted_result = Result.sort_values('Rank')
    
    return sorted_result['Renewable > Median']

task_thirteen()
```

```python
def task_fourteen():

    ContinentDict = {
        'China': 'Asia',
        'United States': 'North America',
        'Japan': 'Asia',
        'United Kingdom': 'Europe',
        'Russian Federation': 'Europe',
        'Canada': 'North America',
        'Germany': 'Europe',
        'India': 'Asia',
        'France': 'Europe',
        'South Korea': 'Asia',
        'Italy': 'Europe',
        'Spain': 'Europe',
        'Iran': 'Asia',
        'Australia': 'Australia',
        'Brazil': 'South America'
    }
    
    Result['Continent'] = Result.index.map(ContinentDict)
    
    sorted = Result.groupby('Continent')['Population'].agg(['size', 'sum', 'mean', 'std'])
    return sorted

task_fourteen()
```

```python
import matplotlib.pyplot as plt

x = Result['Rank']
y = Result['% Renewable']
sizes = Result['2015'] / 0.5e10
continents = Result['Continent']

continent_colors = {
    'Asia': 'red',
    'North America': 'blue',
    'Europe': 'green',
    'Australia': 'yellow',
    'South America': 'orange'
}
colors = continents.map(continent_colors)

plt.figure(figsize=(10, 5))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, edgecolors="w", linewidth=0.5)

for i, country in enumerate(Result.index):
    plt.text(x.iloc[i], y.iloc[i], country, fontsize=6, ha='center', va='center')

plt.title('Bubble chart', fontsize=12)
plt.xlabel('Rank', fontsize=12)
plt.ylabel('% Renewable', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
